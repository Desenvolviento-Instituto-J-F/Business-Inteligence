```python
import pandas as pd
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine    
import streamlit as st
import plotly.express as px
print("CWD:", os.getcwd())
load_dotenv()
print("HOST ENV:", os.getenv("DB_HOST_NAME"))
```

```python
#criar conexao 
def conn():
    load_dotenv(override=True)

    engine = create_engine(
        f"postgresql+psycopg2://{os.getenv('DB_USER')}:{os.getenv('DB_PASS')}"
        f"@{os.getenv('DB_HOST_NAME')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}?sslmode=require"
    )
    return engine

    return engine

#p cor das barras
def status(media):
    if media >=7:
        return 'Acima'
    elif media >=5:
        return 'Na média'
    else:
        return 'Abaixo'

st.title("Nota média por disciplina")

```

```python
#deifinir dados
qntd_disciplinas = pd.read_sql("select count(nome) as total from disciplina",conn()
)
qntd_disciplinas = int(qntd_disciplinas['total'].iloc[0])

df_disc = pd.read_sql(
    "SELECT nome FROM disciplina ORDER BY id_disciplina",
    conn()
)

media_geral = pd.read_sql(
    "select avg(valor) as mg from nota",
    conn()
)
media_geral = round(float(media_geral['mg'].iloc[0]),2)

dados = {
    "disciplina": df_disc["nome"].tolist(),
    "media_geral": media_geral
}

(dados)
```

```python

df = pd.DataFrame(dados)
df['status'] = df["media"].apply(status)

col1, col2, col3 = st.columns(3)

disciplinas = len(df)
media_geral = round(df["media"].mean(),2)
aprovados = (df["media"] >= 6).sum()

with col1:
    st.metric("Disciplinas", disciplinas)

with col2:
    st.metric("Média geral", media_geral)

with col3:
    st.metric("Acima da média", aprovados)


st.subheader("Nota média por disciplina")

fig = px.bar(
    df,
    x="disciplina",
    y="media",
    color="status",
    color_discrete_map={
        "Acima":"#34AD38"
        ,"Na média": "#E3E041"
,"Abaixo":"#FF4D4F"
    }
)


st.plotly_chart(fig, use_container_width=True)

```
